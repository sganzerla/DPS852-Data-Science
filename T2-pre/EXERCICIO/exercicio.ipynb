## 2 - Selecionar os países da América do Sul e persistir o resultado.

```python
import pandas as pd

caminho =  './'
novo_arquivo = 'dataset_bancomundial_modificado.csv'
complemento_arquivo = 'dataset_bancomundial_paises.csv'

cod = ['ARG', 'BRA', 'BOL',  'CHL', 'COL', 'ECU', 'GUY', 'PER', 'PRY', 'SUR', 'URY', 'VEN']

def get_csv(arquivo: str):
    return pd.read_csv(caminho + arquivo, sep = ';', decimal = ',')

def selecionar_colunas(dataframe, lista_colunas: list[str]):
    return dataframe[lista_colunas]

def filtrar_ocorrencias_de_um_termo_em_coluna(dataframe, nome_coluna: str, termo_pesquisado: str):
    return dataframe[dataframe[nome_coluna].str.contains(termo_pesquisado)]

def filtrar_ocorrencias_de_varios_termos_em_coluna(dataframe, nome_coluna: str, termos_pesquisados: list[str]):
    return dataframe.loc[pd.Series(dataframe[nome_coluna].isin(termos_pesquisados))]

dataset_exercicio1 = get_csv(novo_arquivo)
dataset_complemento = get_csv(complemento_arquivo)

# SELECIONANDO APENAS TRES COLUNAS
selecao_paises_america_latina = selecionar_colunas(dataset_complemento, ['Short Name','Country Code', 'Region'])

# COMPLETANDO LINHAS COM 0 ONDE HÁ VALORES NULOS
selecao_paises_america_latina.fillna('0', inplace = True)

# SEPARANDO OCORRENCIAS QUE CONTENHAM LATIN NA COLUNA REGION
paises_latinos = filtrar_ocorrencias_de_um_termo_em_coluna(selecao_paises_america_latina, 'Region', 'Latin')

# FILTRAR PAISES DA AMERICA DO SUL
paises_america_sul = filtrar_ocorrencias_de_varios_termos_em_coluna(paises_latinos, 'Country Code', cod) 

display(paises_america_sul)

# display(dataset_exercicio1)
dataset_filtrado = filtrar_ocorrencias_de_varios_termos_em_coluna(dataset_exercicio1, 'Country Code', paises_america_sul['Country Code'] )
display(dataset_filtrado)
```

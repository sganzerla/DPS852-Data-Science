Criando as tabelas do pequeno negocio e consultando as tabelas criadas:

```python
# from google.colab import drive
import pandas as pd
import numpy as np
import statistics as stat
import matplotlib.pyplot as plt
from datetime import datetime, date, timedelta
import sqlite3

# drive.mount('/content/gdrive')

caminho = './'

nome = 'pequeno_negocio.db'

nomecaminho = caminho+nome

# conecta cria a conexao com o banco de dados
conn = sqlite3.connect(nomecaminho)
print("Conexão aberta!")


tabelas = pd.read_sql_query(
    "SELECT name FROM sqlite_master WHERE type='table'", conn)
print('\n\ntabelas no banco ', nome, ' antes:')
print(tabelas)


# le o comando sql do arquivo para ser executado
nome = 'pequeno_negocio_cria_tabelas.sql'
nomecaminho = caminho+nome
qry = open(nomecaminho, 'r').read()
# print(qry)

# executa o comando para criar as tabelas
conn.executescript(qry)

tabelas = pd.read_sql_query(
    "SELECT name FROM sqlite_master WHERE type='table'", conn)
print('\n\ntabelas no banco ', nome, ' depois:')
print(tabelas)

# transfere as alteracoes para o banco de dados
conn.commit()

# fecha a conexao com o banco de dados
conn.close()
```

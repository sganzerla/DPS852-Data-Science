```python
Buscando as linhas cujas coluna 'Indicator Name' que contenha 'GDP per capita':
```

```python
from google.colab import drive 
import pandas as pd
import numpy as np
from datetime import datetime, date, timedelta

drive.mount('/content/gdrive')

caminho='gdrive/MyDrive/ufsm/aulas/introducao_ciencia_analise_dados_2022_1/aulas_13_16_python/'
#caminho='./'

nome='dataset_bancomundial.csv'

nomecaminho=caminho+nome

# le o dataset cujo nome esta definido na variavel 'nomecaminho'
gfdd = pd.read_csv(nomecaminho, sep=';', decimal=',')

conteudo_coluna='GDP per capita'

gfdd_linhas_gdp = gfdd[ gfdd['Indicator Name'].str.contains(conteudo_coluna) ]
gfdd_linhas_gdp = gfdd_linhas_gdp.reset_index(drop=True)

linhas,colunas=gfdd_linhas_gdp.shape
print('Número de linhas : ',linhas)
print('Número de colunas : ',colunas)


#print('Número de linhas : ', gfdd_linhas_gdp.shape[0])

gfdd_linhas_gdp.head(10)

#print(len(gfdd_linhas_gdp))

#gfdd.columns
```

Assumindo a primeira data como data inicial e a última como data final:

```python
# from google.colab import drive 
import pandas as pd
import numpy as np
from datetime import datetime, date, timedelta

# drive.mount('/content/gdrive')

# caminho='gdrive/MyDrive/ufsm/aulas/introducao_ciencia_analise_dados_2022_1/aulas_9_12_python/'
caminho='./'

dataset='dataset_reg_chuva_santa_maria_2011_2020_2.csv'

caminho_dataset=caminho+dataset

# le o dataset cujo nome esta definido na variavel 'dataset'
precipitacao = pd.read_csv(caminho_dataset, sep=';')

precipitacao=precipitacao.sort_values(by=['DT_MEDICAO', 'HR_MEDICAO'])

precipitacao=precipitacao.reset_index(drop=True)
#print('data set com a coluna de indice agora conforme a nova sequencia')
#print(precipitacao_ord)

datas=precipitacao['DT_MEDICAO']
data_inicial=datas[0]
#print('primeira data:',data_inicial)

data_inicial_date=datetime.strptime(data_inicial, '%Y-%m-%d')
print('data inicial: ', data_inicial_date)

num_linhas=len(precipitacao)
data_final=datas[num_linhas-1]
data_final_date=datetime.strptime(data_final, '%Y-%m-%d')
print('data final: ', data_final_date)

dif_data_final_data_inicial=data_final_date-data_inicial_date

print('numero de dias do periodo: ',dif_data_final_data_inicial.days)
```

```python
# from google.colab import drive 
import pandas as pd
import numpy as np
from datetime import datetime, date, timedelta

# drive.mount('/content/gdrive')

# caminho='gdrive/MyDrive/ufsm/aulas/introducao_ciencia_analise_dados_2022_1/aulas_9_12_python/'
caminho='./'

dataset='dataset_reg_chuva_santa_maria_2011_2020_2.csv'

caminho_dataset=caminho+dataset

# le o dataset cujo nome esta definido na variavel 'dataset'
precipitacao = pd.read_csv(caminho_dataset, sep=';')

precipitacao=precipitacao.sort_values(by=['DT_MEDICAO', 'HR_MEDICAO'])

precipitacao=precipitacao.reset_index(drop=True)
#print('data set com a coluna de indice agora conforme a nova sequencia')
print(precipitacao)

datas=precipitacao['DT_MEDICAO']
data_inicial=datas[0]
#print('primeira data:',data_inicial)

data_inicial_date=datetime.strptime(data_inicial, '%Y-%m-%d')
#print('data inicial: ', data_inicial_date)

num_linhas=len(precipitacao)
data_final=datas[num_linhas-1]
data_final_date=datetime.strptime(data_final, '%Y-%m-%d')
#print('data final: ', data_final_date)

dif_data_final_data_inicial=data_final_date-data_inicial_date

#print('numero de dias do periodo: ',dif_data_final_data_inicial.days)

datas_periodo = pd.date_range(data_inicial_date, periods=(dif_data_final_data_inicial.days+1))
#print(datas_periodo)

#dias_periodo_leituras=pd.

# construindo um array com zeros do tamanho de 'datas_periodo'
leituras=np.zeros(len(datas_periodo),dtype=int)

base={ 'datas': datas_periodo, 'leituras': leituras }

dias_periodo_leituras=pd.DataFrame(base)

print(dias_periodo_leituras)
```

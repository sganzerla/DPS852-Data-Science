```python
from google.colab import drive 
import pandas as pd
import numpy as np
import statistics as stat
import matplotlib.pyplot as plt
from datetime import datetime, date, timedelta

drive.mount('/content/gdrive')

caminho='gdrive/MyDrive/ufsm/aulas/introducao_ciencia_analise_dados_2022_1/aulas_21_24/'
#caminho='./'

nome='gfdd_gdp_paises_america_sul.csv'

nomecaminho=caminho+nome

# le o dataset cujo nome esta definido na variavel 'nomecaminho'
gfdd_gdp_paises_america_sul = pd.read_csv(nomecaminho, sep=';', decimal=',')

#print(gfdd_gdp_paises_america_sul)


nomes_colunas=['COD_PAIS', 'NUM_MED_VALIDAS', 'NUM_MED_AUSENTES', 'MEDIA', 'MINIMO', 'MAXIMO', 'MEDIANA', 'DESVIO_PADRAO', 'VARIANCIA' ]

analise_paises_america_sul=pd.DataFrame([],columns=nomes_colunas)
#print(analise_paises_america_sul)

estats_pais_i = [0] * len(nomes_colunas)


for indice, linha in gfdd_gdp_paises_america_sul.iterrows():
    linha_cont=linha.to_numpy()
    #print(linha_cont)
    pais=linha_cont[2]
    linha_cont = np.delete(linha_cont, 0)
    linha_cont = np.delete(linha_cont, 0)
    linha_cont = np.delete(linha_cont, 0)
    medidas=[float(x) for x in linha_cont]
    num_medidas_total=len(medidas)
    #print(linha_cont)    
    num_medidas_ausentes=np.count_nonzero(np.isnan(medidas))
    num_medidas_validas=num_medidas_total-num_medidas_ausentes
    #print(num_medidas_total,'-',num_medidas_ausentes,'-',num_medidas_validas)
    
    medidas = np.where(np.isnan(medidas), 0, medidas)
    
    estats_pais=estats_pais_i.copy()
    estats_pais[0]=pais
    estats_pais[1]=num_medidas_validas
    estats_pais[2]=num_medidas_ausentes
    estats_pais[3]=stat.mean(medidas)
    estats_pais[4]=min(medidas)
    estats_pais[5]=max(medidas)
    estats_pais[6]=stat.median(medidas)
    estats_pais[7]=stat.stdev(medidas)
    estats_pais[8]=stat.variance(medidas)
    #print(estats_pais)    
    analise_paises_america_sul.loc[ len(analise_paises_america_sul) ]=estats_pais
    
print(analise_paises_america_sul)

nome_saida='estisticas_gdp_america_sul.csv'
nomecaminho_saida=caminho+nome_saida
analise_paises_america_sul.to_csv(nomecaminho_saida, index=False, sep=';')
```

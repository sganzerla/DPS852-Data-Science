Ordenando o dataset e reinicializando a coluna indice, observando que a coluna 'index' original foi mantida:

```python
# from google.colab import drive 
import pandas as pd
import numpy as np
from datetime import datetime, date, timedelta

# drive.mount('/content/gdrive')

# caminho='gdrive/MyDrive/ufsm/aulas/introducao_ciencia_analise_dados_2022_1/aulas_9_12_python/'
caminho='./'

dataset='dataset_reg_chuva_santa_maria_2011_2020_2.csv'

caminho_dataset=caminho+dataset

# le o dataset cujo nome esta definido na variavel 'dataset'
precipitacao = pd.read_csv(caminho_dataset, sep=';')

precipitacao_ord=precipitacao.sort_values(by=['DT_MEDICAO', 'HR_MEDICAO'])

precipitacao_ord=precipitacao_ord.reset_index()

print('data set com a coluna de indice agora conforme a nova sequencia')
print(precipitacao_ord)

```

Ordenando o dataset e reinicializando a coluna indice, observando que a coluna 'index' foi descartada:

```python
# from google.colab import drive 
import pandas as pd
import numpy as np
from datetime import datetime, date, timedelta

# drive.mount('/content/gdrive')

# caminho='gdrive/MyDrive/ufsm/aulas/introducao_ciencia_analise_dados_2022_1/aulas_9_12_python/'
caminho='./'

dataset='dataset_reg_chuva_santa_maria_2011_2020_2.csv'

caminho_dataset=caminho+dataset

# le o dataset cujo nome esta definido na variavel 'dataset'
precipitacao = pd.read_csv(caminho_dataset, sep=';')

precipitacao_ord=precipitacao.sort_values(by=['DT_MEDICAO', 'HR_MEDICAO'])

precipitacao_ord=precipitacao_ord.reset_index(drop=True)

print('data set com a coluna de indice agora conforme a nova sequencia')
print(precipitacao_ord)
```

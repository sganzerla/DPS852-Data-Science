Construindo o dataframe com o indicador 'GDP per capita', excluindo as colunas 'Indicator Name' e 'Indicator Code' e persistindo o resultado no arquivo 'gfdd_gdp.csv':

```python
from google.colab import drive 
import pandas as pd
import numpy as np
from datetime import datetime, date, timedelta

drive.mount('/content/gdrive')

caminho='gdrive/MyDrive/ufsm/aulas/introducao_ciencia_analise_dados_2022_1/aulas_21_24/'
#caminho='./'

nome='dataset_bancomundial.csv'

nomecaminho=caminho+nome

# le o dataset cujo nome esta definido na variavel 'nomecaminho'
gfdd = pd.read_csv(nomecaminho, sep=';', decimal=',')

conteudo_coluna='GDP per capita'

gfdd_linhas_gdp = gfdd[ gfdd['Indicator Name'].str.contains(conteudo_coluna) ]
gfdd_linhas_gdp = gfdd_linhas_gdp.reset_index(drop=True)

#print(gfdd_linhas_gdp)

colunas_para_excluir=[ 'Indicator Name', 'Indicator Code' ]

gfdd_gdp=gfdd_linhas_gdp.drop(colunas_para_excluir, axis=1)

nome_saida='gfdd_gdp.csv'
nomecaminho_saida=caminho+nome_saida
gfdd_gdp.to_csv(nomecaminho_saida, index=False, sep=';')

gfdd_gdp
```

Buscando os pa√≠ses da Am√©rica do Sul do dataset 'dataset_bancomundial_paises.csv' e criando o dataset 'gfdd_paises_america_sul':

```python
from google.colab import drive 
import pandas as pd
import numpy as np
from datetime import datetime, date, timedelta

drive.mount('/content/gdrive')

caminho='gdrive/MyDrive/ufsm/aulas/introducao_ciencia_analise_dados_2022_1/aulas_21_24/'
#caminho='./'

nome='dataset_bancomundial_paises.csv'

nomecaminho=caminho+nome

# le o dataset cujo nome esta definido na variavel 'nomecaminho'
gfdd_paises = pd.read_csv(nomecaminho, sep=';', decimal=',')

gfdd_paises.columns

print(gfdd_paises['Region'].unique())

regiao_intesse='Latin America'

gfdd_paises.fillna( '0', inplace=True )
gfdd_paises_america_latina_caribe = gfdd_paises[ gfdd_paises['Region'].str.contains(regiao_intesse) ]

gfdd_paises_america_latina_caribe = gfdd_paises_america_latina_caribe[['Country Code','Short Name']]

print(gfdd_paises_america_latina_caribe)

codigos=gfdd_paises_america_latina_caribe['Country Code'].to_numpy()

codigos_paises_america_sul = [ 'ARG', 'BOL', 'BRA', 'CHL', 'COL', 'ECU', 'GUY', 'PRY', 'PER', 'SUR', 'URY', 'VEN' ]

vazio=1
for pais in codigos_paises_america_sul:
    sel=gfdd_paises_america_latina_caribe[ gfdd_paises_america_latina_caribe['Country Code'].str.contains(pais) ]
    if( vazio==1 ):
        gfdd_paises_america_sul=sel
        vazio=0
    else:
        gfdd_paises_america_sul= pd.concat([gfdd_paises_america_sul,sel],ignore_index=True)

print(gfdd_paises_america_sul)
    
nome_saida='gfdd_paises_america_sul.csv'
nomecaminho_saida=caminho+nome_saida
gfdd_paises_america_sul.to_csv(nomecaminho_saida, index=False, sep=';')
```

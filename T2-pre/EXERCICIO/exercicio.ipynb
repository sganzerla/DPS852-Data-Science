## 2 - Selecionar os países da América do Sul e persistir o resultado.

```python
import pandas as pd

caminho =  './'
arquivo_banco_mundial = 'dataset_bancomundial.csv'
complemento_arquivo = 'dataset_bancomundial_paises.csv'

cod = ['ARG', 'BRA', 'BOL',  'CHL', 'COL', 'ECU', 'GUY', 'PER', 'PRY', 'SUR', 'URY', 'VEN']

def get_csv(arquivo: str):
    return pd.read_csv(caminho + arquivo, sep = ';', decimal = ',')

def selecionar_colunas(dataframe, lista_colunas: list[str]):
    return dataframe[lista_colunas]

def filtrar_ocorrencias_de_um_termo_em_coluna(dataframe, nome_coluna: str, termo_pesquisado: str):
    return dataframe[dataframe[nome_coluna].str.contains(termo_pesquisado)]

def filtrar_ocorrencias_de_varios_termos_em_coluna(dataframe, nome_coluna: str, termos_pesquisados: list[str]):
    return dataframe.loc[pd.Series(dataframe[nome_coluna].isin(termos_pesquisados))]

dataset = get_csv(arquivo_banco_mundial)
dataset_complemento = get_csv(complemento_arquivo)

# SELECIONANDO APENAS TRES COLUNAS
selecao_paises_america_latina = selecionar_colunas(dataset_complemento, ['Short Name','Country Code', 'Region'])

# COMPLETANDO LINHAS COM 0 ONDE HÁ VALORES NULOS
selecao_paises_america_latina.dropna(inplace = True)

# SEPARANDO OCORRENCIAS QUE CONTENHAM LATIN NA COLUNA REGION
paises_latinos = filtrar_ocorrencias_de_um_termo_em_coluna(selecao_paises_america_latina, 'Region', 'Latin')

# FILTRAR PAISES DA AMERICA DO SUL
paises_america_sul = filtrar_ocorrencias_de_varios_termos_em_coluna(paises_latinos, 'Country Code', cod) 

display(paises_america_sul)

# display(dataset)
dataset_filtrado = filtrar_ocorrencias_de_varios_termos_em_coluna(dataset, 'Country Code', paises_america_sul['Country Code'] )

dataset_filtrado.drop(columns=['index', 'Country Name'], inplace = True)
display(dataset_filtrado)

novo_arquivo_paises = 'paises_america_latina.csv'
dataset_filtrado.to_csv(caminho + novo_arquivo_paises, sep=';', decimal=',', index=False)
```

## 3 - Mostrar as estatísticas sobre o `GDP` para estes países: média, min, max, mediana, desvio padrão e variância.

```python
import pandas as pd

caminho = './'
nome_arquivo = 'paises_america_latina.csv'

paises_america_latina = pd.read_csv(caminho + nome_arquivo, sep = ';', decimal = ',')

def filtrar_ocorrencias_de_um_termo_em_coluna(dataframe, nome_coluna: str, termo_pesquisado: str):
    return dataframe[dataframe[nome_coluna].str.contains(termo_pesquisado)]

def filtrar_ocorrencias_de_varios_termos_em_coluna(dataframe, nome_coluna: str, termos_pesquisados: list[str]):
    return dataframe.loc[pd.Series(dataframe[nome_coluna].isin(termos_pesquisados))]

nome_coluna = 'Indicator Name'
filtro_buscado = 'GDP \(Current USD\)'

# filtrando apenas o indicador de gdp dos paises latino-americanos
paises_america_latina  = filtrar_ocorrencias_de_um_termo_em_coluna(paises_america_latina, nome_coluna, filtro_buscado)

# deltando colunas Indicator Name e indicator code
paises_america_latina = paises_america_latina.drop(['Indicator Code', 'Indicator Name'], axis = 1).reset_index(drop = True)

# separando linhas onde há colunas com valores nulos
linhas_na = paises_america_latina[paises_america_latina.isna().any(axis = 1)]

print("Linhas com colunas com valores nulos: ")
display(linhas_na)


# apago colunas onde há um 
paises_america_latina.dropna(axis = 1, inplace = True)

display(paises_america_latina)


```
